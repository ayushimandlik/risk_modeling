# What is LendingClub?
LendingClub is an online lending platform ‚Äî basically a place on the internet where people can borrow money and others can invest in those loans.

# Imagine this:
Say you need a personal loan to pay off your credit cards.


Instead of going to a traditional bank, you go to LendingClub.


You apply for a loan on their website.


LendingClub checks your credit score, income, etc., and assigns you a risk grade.


Investors (could be regular people or institutions) see your loan request and decide to lend you money.


You get the loan, and you pay it back over time ‚Äî with interest.



# How it works for investors:
They invest in fractions of loans to spread the risk.


They earn interest income from borrowers.


If a borrower defaults, investors might lose some money ‚Äî so choosing the right loans is important.



# Why it matters in data science:
LendingClub publishes a huge dataset of historical loans.


It‚Äôs popular for loan default prediction projects because it includes borrower details (income, purpose, credit score, etc.) and whether they repaid or defaulted.


It‚Äôs great for applying machine learning, risk modeling, and explainability tools like SHAP.

# Taking care of outliers!

# Dealing With the Variety of Data Types
The dataset had all sorts of columns:
Numerical features like loan_amnt, revol_bal, dti


Categorical features like home_ownership, verification_status


High-cardinality categories like emp_title, address


Dates like issue_d and earliest_cr_line


To keep things tidy and understandable, I created separate preprocessing functions for:
Numerical features ‚Äî handled missing values, removed/skewed outliers using IQR clipping (only on right-tailed values!), and scaled them.


Categorical features ‚Äî used one-hot encoding for low-cardinality ones.


High-cardinality features ‚Äî frequency encoding or clustering (like for employment titles).


Datetime features ‚Äî calculated meaningful durations like account age, instead of raw dates.


```python
def process_numerical(df, cols, outlier_clip=False):
    df = df.copy()
    for col in cols:
        # Fill missing values with median
        median = df[col].median()
        df[col] = df[col].fillna(median)

        # Clip outliers using IQR
        if outlier_clip:
            q1 = df[col].quantile(0.25)
            q3 = df[col].quantile(0.75)
            iqr = q3 - q1
            lower = q1 - 1.5 * iqr
            upper = q3 + 1.5 * iqr
            df[col] = df[col].clip(lower, upper)
    return df
```

```python
def process_low_card_categoricals(df, cols):
    df = df.copy()
    return pd.get_dummies(df, columns=cols, drop_first=True)
```

```python
def process_high_card_categoricals(df, cols):
    df = df.copy()
    for col in cols:
        freq_map = df[col].value_counts(normalize=True).to_dict()
        df[col] = df[col].map(freq_map)
    return df
```

```python
def process_datetime_features(df, date_cols):
    df = df.copy()
    for col in date_cols:
        df[col] = pd.to_datetime(df[col], errors='coerce')
        df[f"{col}_year"] = df[col].dt.year
        df[f"{col}_month"] = df[col].dt.month

    # If you have at least 2 date columns, create age difference features
    if len(date_cols) >= 2:
        base = date_cols[0]
        for col in date_cols[1:]:
            df[f"{base}_minus_{col}_months"] = (
                (df[base] - df[col]) / np.timedelta64(1, 'M')
            ).round()

    # Drop original datetime columns
    df = df.drop(columns=date_cols)
    return df
```

```python
def process_numerical_right_skewed(df, cols, clip_pct=0.99):
    df = df.copy()
    for col in cols:
        median = df[col].median()
        df[col] = df[col].fillna(median)
        upper = df[col].quantile(clip_pct)
        df[col] = df[col].clip(upper=upper)
    return df
```

```python
cols_to_check = num_cols
outlier_check = df[cols_to_check].describe(
                percentiles=[.01, .25, .5, .75, .99]).T
outlier_check['skewness_check'] = outlier_check['max'] / outlier_check['99%'] 
outlier_check
```

```python
highly_skewed_cols = outlier_check[outlier_check['skewness_check'] > 5].index.to_list()
```

```python
num_cols_no_skew = set(num_cols) - set(highly_skewed_cols)

df = process_numerical(df, num_cols_no_skew)
df = process_numerical_right_skewed(df, highly_skewed_cols, clip_pct=0.99)
df = process_datetime_features(df, dates)
df = process_high_card_categoricals(df, high_card_cats)
df = process_low_card_categoricals(df, low_card_cats)
```

Grid search is like speed dating for your model‚Äôs hyperparameters ‚Äî it tries out all the possible combinations until it finds the one that clicks best! You set up a menu of options (like different learning rates, tree depths, or regularization values), and grid search runs through every combo, checking how each performs. It might take a while, especially with complex models, but it‚Äôs a solid way to squeeze out extra performance and make sure your model is living its best life. Think of it as giving your model a personal stylist ‚Äî and the results are usually worth the wait!

```python
def plot_learning_evolution(r):
    plt.figure(figsize=(12, 8))
    
    plt.subplot(2, 2, 1)
    plt.plot(r.history['loss'], label='Loss')
    plt.plot(r.history['val_loss'], label='Val. Loss')
    plt.title('Loss evolution during trainig')
    plt.legend()

    plt.subplot(2, 2, 2)
    plt.plot(r.history['AUC'], label='AUC')
    plt.plot(r.history['val_AUC'], label='Val. AUC')
    plt.title('AUC score evolution during trainig')
    plt.legend()

def nn_model(num_columns, num_labels, hidden_units, dropout_rates, learning_rate):
    inp = Input(shape=(num_columns, ))
    x = BatchNormalization()(inp)
    x = Dropout(dropout_rates[0])(x)
    for i in range(len(hidden_units)):
        x = Dense(hidden_units[i], activation='relu')(x)
        x = BatchNormalization()(x)
        x = Dropout(dropout_rates[i + 1])(x)
    x = Dense(num_labels, activation='sigmoid')(x)
  
    model = Model(inputs=inp, outputs=x)
    model.compile(optimizer=Adam(learning_rate), loss='binary_crossentropy', metrics=[AUC(name='AUC')])
    return model

num_columns = X_train.shape[1]
num_labels = 1
hidden_units = [150, 150, 150]
dropout_rates = [0.1, 0, 0.1, 0]
learning_rate = 1e-3

model = nn_model(
    num_columns=num_columns, 
    num_labels=num_labels,
    hidden_units=hidden_units,
    dropout_rates=dropout_rates,
    learning_rate=learning_rate
)
r = model.fit(
    X_train, Y_train,
    validation_data=(X_test, Y_test),
    epochs=10,
    batch_size=128
)
```

```python
from sklearn.model_selection import KFold

kfold = KFold(n_splits=5, shuffle=True, random_state=42)

# To store results
fold_accuracies = []

for fold, (train_idx, val_idx) in enumerate(kfold.split(X_train), 1):
    print(f"\nTraining on Fold {fold}...")

    # Split data
    X_train_fold, X_val_fold = X_train[train_idx], X_train[val_idx]
    Y_train_fold, Y_val_fold = Y_train[train_idx], Y_train[val_idx]
    model = nn_model(num_columns=num_columns, 
                     num_labels=num_labels,
                     hidden_units=hidden_units,
                     dropout_rates=dropout_rates,
                     learning_rate=learning_rate)

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['AUC'])

    # Save best model per fold
    weight_save_callback = ModelCheckpoint(
        f'fold_{fold}_weights.keras',
        monitor='val_loss',
        save_best_only=True,
        verbose=2
    )

    # Fit the model
    history = model.fit(
        X_train_fold, Y_train_fold,
        validation_data=(X_val_fold, Y_val_fold),
        epochs=15,
        batch_size=64,
        callbacks=[weight_save_callback],
        verbose=1)

    # Evaluate
    scores = model.evaluate(X_val_fold, Y_val_fold, verbose=0)
    print(f"Fold {fold} Validation Accuracy: {scores[1]:.4f}")
    fold_accuracies.append(scores[1])

print("\nAverage Validation Accuracy Across Folds:", np.mean(fold_accuracies))
```

```python
def get_test_train(df):
    train, test = train_test_split(df, test_size=0.3, random_state=42)

    Y_train = train['default']
    X_train = train.drop('default', axis=1)
    Y_test = test['default']
    X_test = test.drop('default', axis=1)

    X_train = np.array(X_train).astype(np.float32)
    X_test = np.array(X_test).astype(np.float32)
    Y_train = np.array(Y_train).astype(np.float32)
    Y_test = np.array(Y_test).astype(np.float32)
    
    return X_train, X_test, Y_train, Y_test
```
